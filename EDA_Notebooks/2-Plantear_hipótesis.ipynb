<h2>Importamos librerias:</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

<h2>Cargamos el Dataset limpio:</h2>

```python
df1 = pd.read_csv("CSVs/dataframe spotify 2020-21 (Limpio).csv", encoding = "latin1", sep = ",", index_col = 0)
pd.set_option("display.max_rows",10, "display.max_columns", None)
df1
```

<h2>Pasamos de milisegundos a minutos la columna de "Duration":</h2>

```python
#Renombramos columna:
df1 = df1.rename(columns={'Duration (ms)':'Duration'})

#Convertimos los str() a float():
df1['Duration'] = pd.to_numeric(df1.Duration, errors='coerce')

#Aplicamos una función lambda que divide todos los valores de la columna entre 60000:
df1['Duration'] = df1['Duration'].map(lambda c: round((c / 60000),2))
```

<h2>Sacamos un gráfico de correlaciones para poder sacar hipótesis:</h2>

```python
plt.figure(figsize=(15,15))
sns.heatmap(df1.corr(),
            vmin=-1,
            vmax=1,
            cmap=sns.diverging_palette(145, 280, s=85, l=25, n=7),
            square=True,
            linewidths=.1,
            annot=True);
```
