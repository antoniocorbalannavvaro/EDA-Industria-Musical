```python
df2['Song Name'] = df2['Song Name'].map(lambda c: str(c).lower())
df2['Artist'] = df2['Artist'].map(lambda c: str(c).lower())
```

```python
songs_lower = []
for i in songs:
    songs_lower.append(i.lower())
```

```python
datos1 = df2.loc[df2['Song Name'] == songs_lower[0]]
datos2 = df2.loc[df2['Song Name'] == songs_lower[1]]
datos3 = df2.loc[df2['Song Name'] == songs_lower[2]]
datos4 = df2.loc[df2['Song Name'] == songs_lower[3]]
datos5 = df2.loc[df2['Song Name'] == songs_lower[4]]
datos6 = df2.loc[df2['Song Name'] == songs_lower[5]]
datos7 = df2.loc[df2['Song Name'] == songs_lower[6]]
datos8 = df2.loc[df2['Song Name'] == songs_lower[7]]
datos9 = df2.loc[df2['Song Name'] == songs_lower[8]]
datos10 = df2.loc[df2['Song Name'] == songs_lower[9]]

df_final = pd.concat([datos1,datos2,datos3,datos4,datos5,datos6,datos7,datos8,datos9,datos10], ignore_index=True)
```
