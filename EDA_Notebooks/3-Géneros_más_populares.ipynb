<h2>Importamos librerias:</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h2>Cargamos Dataframe:</h2>

```python
df2 = pd.read_csv("CSVs/Top Spotify songs from 2010-2019 - BY YEAR.csv", encoding = "latin1", index_col = 0)
pd.set_option("display.max_rows",10, "display.max_columns", None)
df2
```

<h2>Pasamos la duración a minutos</h2>

```python
df2["dur"] =df2["dur"].map(lambda c: round((c / 60),2))
```

```python
df2["dur"].mean()
```

<h2>Comprobamos que no haya valores nulos ni NaN:</h2>

```python
print(df2.isna().values.any())
print(df2.isnull().values.any())
```

```python
df2.info()
#No hay ;)
```

<h2>Cogemos las columnas que vamos a necesitar:</h2>

```python
df1 = df2[["year","top genre","pop","dur","dB"]]
df1
```

<h2>Dividimos por años el Dataframe:</h2>

```python
mask =(df1["year"].values == 2010)
año_2010 = df1[mask] 

mask =(df1["year"].values == 2011)
año_2011 = df1[mask] 

mask =(df1["year"].values == 2012)
año_2012 = df1[mask] 

mask =(df1["year"].values == 2013)
año_2013 = df1[mask] 

mask =(df1["year"].values == 2014)
año_2014 = df1[mask] 

mask =(df1["year"].values == 2015)
año_2015 = df1[mask] 

mask =(df1["year"].values == 2016)
año_2016 = df1[mask]

mask =(df1["year"].values == 2017)
año_2017 = df1[mask] 

mask =(df1["year"].values == 2018)
año_2018 = df1[mask] 

mask =(df1["year"].values == 2019)
año_2019 = df1[mask] 
```

<h2>Sacamos el género más popular de cada año:</h2>

```python
#Fuentes para los labels:
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```
