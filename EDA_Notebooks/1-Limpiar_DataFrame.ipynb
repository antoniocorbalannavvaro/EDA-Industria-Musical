<h2>Importamos librerias:</h2>

```python
import pandas as pd
```

<h2>Cargamos Dataset:</h2>

```python
df1 = pd.read_csv("CSVs/Spotify Top 200 Charts (2020-2021).csv", encoding = "latin1", sep = ",", index_col = 0)
pd.set_option("display.max_rows",10, "display.max_columns", None)
df1
```

<h2>Cogemos sólo las columnas que necesitamos:</h2>

```python
df1 = df1[["Artist","Song Name","Streams","Release Date","Genre","Loudness","Popularity","Danceability","Tempo","Energy","Chord","Duration (ms)"]]
df1
```

<h2>Cambiamos comas de la columna "Streams" por nada para convertirla a int():</h2>

```python
#Cambiamos comas de la columna "Streams" por nada para convertirla a int()
df1[["Streams"]] = df1[["Streams"]].replace(",","", regex=True)
```

<h2>Sustituimos los NaN por ceros en las columnas de "Tempo" y "Danceability":</h2>

```python
df1 ['Tempo'] = df1 ['Tempo']. fillna(0)
df1 ['Danceability'] = df1 ['Danceability']. fillna(0)
```

<h2>Realizamos las conversiones de "object" a int() o float(), entre otras:</h2>

```python
#df1["Tempo"] = df1["Tempo"].round(0)
```

```python
df1['Artist'] = df1['Artist'].astype(str)
```

```python
df1['Song Name'] = df1['Song Name'].astype(str)
```

```python
df1[["Streams"]] = df1[['Streams']].astype(int)
```

```python
df1["Popularity"] = pd.to_numeric(df1.Popularity, errors='coerce')
```

```python
df1["Loudness"] = pd.to_numeric(df1.Loudness, errors='coerce')
```

```python
df1["Danceability"] = pd.to_numeric(df1.Danceability, errors='coerce')
```

```python
df1["Tempo"] = pd.to_numeric(df1.Tempo, errors='coerce')
```

```python
df1["Energy"] = pd.to_numeric(df1.Energy, errors='coerce')
```

```python
df1['Tempo'] = df1['Tempo'].astype(int)
```
